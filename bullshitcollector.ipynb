```python
import glob
print(glob.glob("C:\\Users\\Honey\\sort-google-scholar\\*.csv"))
csv_files = glob.glob("C:\\Users\\Honey\\sort-google-scholar\\*.csv")
csv_name = []
for i in csv_files:
    csv_name.append(i[35:len(i)])
print(csv_name)
```

```python
import pandas as pd

KEYWORD = ["memory", "homomorphic", "tfhe", "ckks", "pim"]

result_file = open("result.cs", "wt", encoding="UTF-8")

for current_csv in csv_name:
    print(current_csv)
    pdata = pd.read_csv(current_csv)
    title = pdata.Title.to_list()
    author = pdata.Author.to_list()
    target_index = []
    
    for i in range(0, len(title)):
        discard = True
        for kw in KEYWORD:
            if kw in title[i]:
                discard = False
        
        if not discard:
            target_index.append(i)
    print(target_index)

    if len(target_index) != 0:
        result_file.write(current_csv + ",\n")
        for i in target_index:
            result_file.write(author[i].replace(",", "-")+","+title[i]+"\n")


result_file.close()
    
```
